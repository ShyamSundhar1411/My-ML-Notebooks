## *Create a vector, scalar, matrix and tensor with values of your choosing using tf.constant().*

```python
tensor1 = tf.constant([10,7])
tensor1.shape,tf.rank(tensor1),tf.size(tensor1)
```

```python
tensor2 = tf.constant([[[
    [10,7,8],
    [9,12,3],
]]])
tensor2.shape,tf.rank(tensor2),tf.size(tensor2)
```

## *Create two tensors containing random values between 0 and 1 with shape [5, 300].*

```python
generator = tf.random.Generator.from_seed(42)
tensor3 = generator.uniform(shape = (5,300))
tensor4 = generator.uniform(shape = (5,300))
```

```python
tensor3
```

```python
tensor4
```

```python
## *Multiply the two tensors you created in 3 using matrix multiplication*
```

```python
tf.matmul(tensor3,tf.transpose(tensor4))
```

```python
tf.tensordot(tensor3,tf.transpose(tensor4),axes = 0)
```
