## Linear Regression through function

```python
from sklearn.datasets import make_regression
x,y = make_regression(n_samples = 400,n_features = 1,n_informative = 1,noise = 1.8,random_state = 11)
```

```python
x = (x-x.mean())/x.std()
```

```python
x
```

```python
ones = np.ones((x.shape[0],1))
x_ = np.hstack((x,ones))
```

```python
x_
```

```python
def predict(x,theta):
  return np.dot(x,theta)
def getClosedForm(x,y):
  p1 = np.dot(x.T,x)
  p2 = np.dot(x.T,y)
  theta = np.dot(np.linalg.pinv(p1),p2)
  return theta
```

```python
theta = getClosedForm(x_,y)
```

```python
theta
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure()
plt.scatter(x,y)
plt.plot(x,predict(x_,theta),color = "red")
plt.title("Normalized Data")
plt.show()
```

## Linear Regression through Class

```python
class LinearRegression():
  def normalize(self):
    self.X = (self.X-self.X.mean())/self.X.std()
  def find_theta(self):
    p1 = np.dot(self.X.T,self.X)
    p2 = np.dot(self.X.T,self.Y)
    return np.dot(np.linalg.pinv(p1),p2)
  def process(self,x):
    ones = np.ones((x.shape[0],1))
    return np.hstack((x,ones))
  def fit(self,X,Y):
    self.X = self.process(X)
    self.Y = Y.reshape(-1,1)
    self.theta = self.find_theta()
    return self
  def predict(self,x):
    x = self.process(x)
    return np.dot(x,self.theta)
  

```

```python
model = LinearRegression()
model.fit(x,y)
model.normalize()
```

```python
model.theta
```

```python
plt.figure()
plt.scatter(x,y)
plt.plot(x,model.predict(x),color = "red")
plt.title("Normalized Data")
plt.show()
```
